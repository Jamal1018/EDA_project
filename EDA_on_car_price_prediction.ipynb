# **Data Cleaning**

Data Cleaning is the major process in EDA analysis where the below operations will perform:

- Removing/merging relevant columns
- Sorting all the basic issues like lower/uppercase handling, removing special irrelevant characters
-changing the datatypes of the columns as per the syntax
-handling the missing values:







```python
#Creating a copy the dataset 
df = used_car.copy()
```

```python
#checking for duplicates
df[df.duplicated()].shape
```

```python
#droping the duplicates and assigning the operation as permanent
df.drop_duplicates(inplace = True)
```

```python
df[df.duplicated()].shape
```

```python
#observation after removing the duplicate rows
df.shape
```

```python
#removing the special character '₹' from the 'price ' column to convert this into float datatype
df['car_price_in_rupees'] = df['car_price_in_rupees'].str.replace('₹','').str.strip().str.split(' ').str.get(0)
```

```python
df['car_price_in_rupees'].str.strip().str.replace(',','')
```

 In row(1777) the price has inccurate value of 1.36 lakhs which has to replace with 136 lakhs accuracy*

```python
#finding the inaccurate error in car_price_in_rupees and reassigning its value
df[df['car_price_in_rupees'] == '1.36']
```

```python
df.loc[1850]
```

```python
df.loc[1850, 'car_price_in_rupees'] = 136 
```

```python
#converting the string to float datatype 
df['car_price_in_rupees'] = df['car_price_in_rupees'].astype(float)
```

```python
#Finding the particular rows which having the string type and sorting the error 
df[df['car_price_in_rupees'] == '99,991']
```

```python
df.loc[1635,'car_price_in_rupees'] = 0.99
```

```python
df[df['car_price_in_rupees'] == '60,000']
```

```python
df.loc[499,'car_price_in_rupees'] = 0.60
```

```python
df.info()
```

###Multiplying the car price column with '100000' to get accurate results with this column

```python
df['car_price_in_lakhs'] = df['car_price_in_lakhs'] * 100000
```

```python
#replacing the special character of 'km' in kms_driven column and converting the datatype to integer
df['kms_driven'] = df['kms_driven'].str.replace('km','').str.replace(',','').astype(int)
```

```python
#Creating the new column 'brand name' by using the 'car_name' column
brand = df['car_name'].str.strip().str.split(' ').str.get(0)
```

```python
#creating a new column for each brand of the car 
df.insert(1,'brand_name', brand)
```

```python
df
```

```python
#using lambda function to lower the case of alphabets
df['brand_name'] = df['brand_name'].apply(lambda x : x.lower())
```

```python
df['car_name'] = df['car_name'].apply(lambda x : x.lower())
```

```python
#renaming the 'car_price_in_rupees' column to ' car_price_in_lakhs' to represent its actual value
df.rename(columns={'car_price_in_rupees' : 'car_price_in_lakhs'}, inplace = True)
```

```python
#checking the fuel type categories
df['fuel_type'].value_counts()
```

###Adding the same fuel types of the car as the petrol , petrol +1 representing the same fuel type in the car simulatneously with the diesel fuel type

```python
#adding the fuel types
temp_df = df[df['fuel_type'] == 'Petrol + 1']
```

```python
df.loc[temp_df.index,'fuel_type'] ='Petrol'
```

```python
temp_df = df[df['fuel_type'] == 'Diesel + 1']
```

```python
df.loc[temp_df.index,'fuel_type'] ='Diesel'
```

```python
#overall info of the dataset 
df.describe()
```

```python
df = df.reset_index()
```

```python
df['index'] = df['index'] +1
```
